```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("insurance.csv")
```

```python
df.head()
```

<h2> Finding The Age Distribution

```python
df.shape
```

```python
plt.hist(df['charges'], bins=30, color='skyblue', edgecolor='black')
plt.title("Distribution of Insurance Charges")
plt.xlabel("Charges")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
plt.hist(df['children'], bins=6,color='lightgreen', edgecolor='black')
plt.title("Children Distribution")
plt.xlabel("Number of Children")
plt.ylabel("Frequency")
plt.show()
```

<h2> Region Wise People

```python
region_counts = df['region'].value_counts()

plt.barh(region_counts.index, region_counts.values,
         color='skyblue', edgecolor='black')

plt.title("Number of Customers by Region")
plt.xlabel("Count")
plt.ylabel("Region")
plt.grid(True)


plt.show()
```

<h2>Smoker Distribution

```python
smoker_counts = df['smoker'].value_counts()

plt.barh(smoker_counts.index, smoker_counts.values,
         color='orange', edgecolor='black')

plt.title("Smoker vs Non-Smoker")
plt.xlabel("Count")
plt.ylabel("Category")
plt.grid(True)

plt.show()
```
